```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.head()
```

Changing format and dtype

```python
df['Year'] = df['Date'].dt.year
df['Runs']=df['Runs'].astype(str)
```




Creating new column from runs status(out/not out)




```python
df['Out/Not-Out']=df['Runs'].str.contains(r'[^\w\s]', regex=True)
df.loc[df['Out/Not-Out'] == True, 'Out/Not-Out'] = 'Not Out' 
df.loc[df['Out/Not-Out'] == False, 'Out/Not-Out'] = 'Out'
```

changing Runs format from string to integer

```python
df['Runs'] = df['Runs'].str.replace(r'[^\w\s]', '')
df['Runs']=df['Runs'].astype(str).astype(int)
```

```python

label = df['Against'].value_counts()

# Create explode list dynamically (explode only the first slice)
explode = [0.1] + [0] * (len(label) - 1)
colors = [
    '#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
    '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf'
]
plt.figure(figsize=(20, 10))
plt.pie(
    label,
    labels=label.index,
    autopct='%1.2f%%',
    textprops={'fontsize': 9},
    shadow=True,
    explode=explode,
    colors=colors[:len(label)]
)
plt.title('Most Centuries Against Pie Chart', fontsize=20, pad=10)
plt.legend(bbox_to_anchor=(1, 1.01))
plt.show()
```
